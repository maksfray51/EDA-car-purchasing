## Main idea.


Let's say the customers in this database want to lease or loan a car. We will do some analysis to decide whether to approve a loan to buy a car or not.

We will do this based on the following indicators:
1.   The ratio of their current credit card debt to their net worth does not exceed 0.3 
2.   The ratio of annual salary to credit card debt is at least 6
3.   The ratio of the annual salary to the cost of the car is not less than 2



Here are listed all the customers who are one hundred percent likely to have a loan approved

````python
df.loc[(df.annual_salary / df.car_purchase_amount >= 2) & (df.ratio_of_credit_card_debt_to_net_worth <= 0.3) & (df.annual_salary / df.credit_card_debt >= 6)]
````

## **Linear regression**

Let's build a model that will predict the cost of a purchasing car that a customer may want based on its net worth

````python
import numpy as np
````

Preparing our data

````python
from sklearn.linear_model import LinearRegression

# Here we use reshape to make our array two-dimensional
y = df.car_purchase_amount.values
x = df.net_worth.values.reshape(-1,1)
````

Building a model and extract the coefficients of a straight line, which will be a linear regression

````python
# f(x) = b₀ + b₁x
model = LinearRegression().fit(x, y)

r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)

# b₀
print('intercept:', model.intercept_)

# b₁
print('slope:', model.coef_)
````

Now, let's build a test array of net worth that we will use to predict the value of the purchasing car

````python
# our test values
rng = np.random.default_rng()
x_new = rng.integers(low=40000, high=1000000, size=(100, 1))

# Our predicted values
y_predicted = model.predict(x_new)
print('predicted response:\n', y_predicted)
````

### Visualization of linear regression

````python
from sklearn import linear_model

######################################## Data preparation #########################################
X = df.net_worth.values.reshape(-1,1)
y = df.car_purchase_amount.values

################################################ Train #############################################

ols = linear_model.LinearRegression()
model = ols.fit(X, y)
response = model.predict(X)

############################################## Evaluate ############################################

r2 = model.score(X, y)

############################################## Plot ################################################

plt.style.use('default')
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(X, response, color='k', label='Regression model')
ax.scatter(X, y, edgecolor='k', facecolor='grey', alpha=0.7, label='Sample data')
ax.set_ylabel('Car purchasing amount', fontsize=14)
ax.set_xlabel('Net worth', fontsize=14)
ax.legend(facecolor='white', fontsize=11)
ax.set_title('$R^2= %.2f$' % r2, fontsize=18)

ax.text(0.65, 0.1, '$y = %.3f x_1 + %.3f $' % (model.coef_, model.intercept_),
             fontsize=12, transform=ax.transAxes)

fig.tight_layout()
````

## Clustering

Just visualasing our source data

````python
from sklearn.cluster import KMeans

X = df.net_worth.values.reshape(-1, 1)
y = df.car_purchase_amount.values

plt.scatter(X,y, label='True Position')
````

Let's cluster!

````python
kmeans = KMeans(n_clusters=3) 
kmeans.fit(X, y)

# Centers of our clusters
print(kmeans.cluster_centers_)

# Here we can see what cluster has each point of our data
print(kmeans.labels_)
````

Let's Visualize it 

````python
plt.scatter(X, y, c=kmeans.labels_, cmap='rainbow') 
````
